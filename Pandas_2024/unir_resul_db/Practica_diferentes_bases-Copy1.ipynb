## 5. Ejemplo completo

```python
from sqlalchemy import create_engine
from pymongo import MongoClient
import pandas as pd

# Conexi√≥n a SQL Server
engine_sql = create_engine('mssql+pyodbc://username:password@server/database?driver=SQL+Server')
employees_sql = pd.read_sql_query('SELECT * FROM Employees', engine_sql)
customers_sql = pd.read_sql_query('SELECT * FROM Customers', engine_sql)
transactions_sql = pd.read_sql_query('SELECT * FROM Transactions', engine_sql)

# Conexi√≥n a MySQL
engine_mysql = create_engine('mysql+pymysql://username:password@host:port/database')
employees_mysql = pd.read_sql_query('SELECT * FROM Employees', engine_mysql)
customers_mysql = pd.read_sql_query('SELECT * FROM Customers', engine_mysql)
transactions_mysql = pd.read_sql_query('SELECT * FROM Transactions', engine_mysql)

# Conexi√≥n a MongoDB
client = MongoClient('mongodb://localhost:27017/')
db = client['TelecomDB']
employees_mongo = pd.DataFrame(list(db.Employees.find()))
customers_mongo = pd.DataFrame(list(db.Customers.find()))
transactions_mongo = pd.DataFrame(list(db.Transactions.find()))
employees_mongo.drop(columns=['_id'], inplace=True)
customers_mongo.drop(columns=['_id'], inplace=True)
transactions_mongo.drop(columns=['_id'], inplace=True)

# Unir DataFrames
employees_combined = pd.concat([employees_sql, employees_mysql, employees_mongo], ignore_index=True)
customers_combined = pd.concat([customers_sql, customers_mysql, customers_mongo], ignore_index=True)
transactions_combined = pd.concat([transactions_sql, transactions_mysql, transactions_mongo], ignore_index=True)

# Guardar en CSV
employees_combined.to_csv('combined_employees.csv', index=False)
customers_combined.to_csv('combined_customers.csv', index=False)
transactions_combined.to_csv('combined_transactions.csv', index=False)
```
